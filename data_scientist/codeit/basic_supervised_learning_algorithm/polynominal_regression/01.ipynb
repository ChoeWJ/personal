```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("../data/boston.csv")
data
```

```python
polynomial_transformer = PolynomialFeatures(2)
```

```python
polynomial_data = polynomial_transformer.fit_transform(data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']])
```

```python
polynomial_data
```

```python
polynomial_data.shape
```

```python
X = pd.DataFrame(polynomial_data, columns=polynomial_transformer.get_feature_names_out())
```

```python
X
```

```python
y = data[['MEDV']]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_test_prediction = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_test_prediction) ** 0.5
```
