```python
order_df['days_of_shipping(actual)'] = order_df['shipping_time'] - order_df.index
```

```python
order_df['days_of_shipping(expected)'] = pd.to_timedelta(order_df['days_for_shipping(expected)'], unit='D')
```

```python
order_df['late_shipping'] = order_df['days_of_shipping(expected)'] < order_df['days_of_shipping(actual)']
```
