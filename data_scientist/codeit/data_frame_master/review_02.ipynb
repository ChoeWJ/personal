```python
import numpy as np
import pandas as pd
```

```python
airbnb_df = pd.read_csv('data/airbnb.csv')
airbnb_df
```

```python
airbnb_df[airbnb_df.isna().any(axis=1)]
```

```python
rating_mean = airbnb_df['rating'].mean()
rating_mean
```

```python
airbnb_df['rating'] = airbnb_df['rating'].fillna(rating_mean)
airbnb_df
```

```python
airbnb_df.isna().sum()
```

```python
airbnb_df[airbnb_df.duplicated()]
```

```python
airbnb_df[airbnb_df.duplicated(subset='id')]
```

```python
airbnb_df = airbnb_df.drop_duplicates(subset='id', keep='last')
airbnb_df
```

```python
q1 = airbnb_df['price'].quantile(0.25)
q3 = airbnb_df['price'].quantile(0.75)

iqr = q3 - q1

lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr

condition1 = airbnb_df['price'] <= upper_limit
condition2 = airbnb_df['price'] >= lower_limit

airbnb_df = airbnb_df[condition1 & condition2]
```
