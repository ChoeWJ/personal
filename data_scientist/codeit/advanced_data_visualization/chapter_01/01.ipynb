```python
birth_df = pd.read_csv("../data/birth_rate.csv")
birth_df
```

```python
birth_25_29 = birth_df.query('연령대 == "25-29세"')
birth_30_34 = birth_df.query('연령대 == "30-34세"')
```

```python
fig, ax = plt.subplots()

ax.plot(birth_25_29['시점'], birth_25_29['여성 천명당 출생아수'], label='25~29세', linewidth=2, linestyle='--')
ax.plot(birth_30_34['시점'], birth_30_34['여성 천명당 출생아수'], label='30~34세', color='skyblue', marker='o')
ax.legend()

ax.set_title('여성 천명당 출생아수: 연령대별 추이')
ax.set_xlabel('연도')
ax.set_ylabel('여성 천명당 출생아수')

ax.set_xticks(birth_25_29['시점'], labels=[f'{i}년' for i in birth_25_29['시점']], rotation=30)
ax.set_yticks(np.arange(0, 110, 10))

ax.spines[['top', 'right']].set_visible(False)
# ax.spines.top.set_visible(False)
# ax.spines.right.set_visible(False)

ax.grid(axis='y', linestyle=':', color='lightgrey')

plt.show()
```

```python
idol_df = pd.read_csv('../data/kpop_idols.csv')
idol_df
```

```python
boy_group_career = idol_df.query("Gender == 'Boy'")
girl_group_career = idol_df.query("Gender == 'Girl'")

followers_mean = idol_df['Followers'].mean()
```

```python
plt.rcParams["figure.figsize"] = (12, 6)
fig, ax = plt.subplots()

ax.scatter(girl_group_career['Career'], girl_group_career['Followers'], alpha=0.5, marker='*', color='coral', label='girl group')
ax.scatter(boy_group_career['Career'], boy_group_career['Followers'], alpha=0.5, marker='p', color='lightseagreen', label='boy group')

ax.set_title("Followers per Career Length")
ax.set_xlabel('Years of Career')
ax.set_ylabel('Instagram Followers (in Millions)')

ax.set_xticks(np.arange(0, 18, 1))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: int(x / 1e6)))

ax.spines[['top', 'right']].set_visible(False)
ax.axhline(y=followers_mean, color='grey', linestyle='--', linewidth=0.5, label='mean')


plt.legend()
plt.show()
```

```python
kr_pisa = pisa_df.query("Country == 'South Korea'")
us_pisa = pisa_df.query("Country == 'United States'")
kr_math = kr_pisa.query("Subject == 'Maths'")
kr_science = kr_pisa.query("Subject == 'Science'")
us_math = us_pisa.query("Subject == 'Maths'")
us_science = us_pisa.query("Subject == 'Science'")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(kr_math['Year'], kr_math['Score'], label='South Korea', color='cadetblue', marker='o')
axes[0].plot(us_math['Year'], us_math['Score'], label='United States', color='salmon', marker='o')
axes[0].set_title("Math Score")
axes[0].legend()

axes[1].plot(kr_math['Year'], kr_science['Score'], color='cadetblue', marker='o')
axes[1].plot(us_math['Year'], us_science['Score'], color='salmon', marker='o')
axes[1].set_title("Science Score")

for ax in axes:
    ax.set_xlabel('Year')
    ax.set_ylabel('Score')
    ax.set_xticks([2006, 2009, 2012, 2015, 2018])
    ax.spines.top.set_visible(False)


plt.show()
```
