````python
def get_book_title(isbn):
    # Yes 24 도서 검색 페이지 URL
    url = 'https://www.yes24.com/Product/Search?domain=BOOK&query={}'
    
    # URL에 ISBNㅇ르 넣어 HTML을 가져온다
    response = requests.get(url.format(isbn))
    soup = BeautifulSoup(response.text, 'html.parser')  # html 파싱
    
    # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져온다
    title = soup.find('a', attrs={'class' : 'gd_name'}).get_text()
    return title
    
get_book_title(9788942340262)
    
````

````python
# import re

# def get_book_info(row):
#     title = row['도서명']
#     author = row['저자']
#     pub = row['출판사']
#     year = row['발행년도']
    
#     # Yes 24 도서 검색 페이지 URL
#     url = 'https://www.yes24.com/Product/Search?domain=BOOK&query={}'
    
#     # URL에 ISBNㅇ르 넣어 HTML을 가져온다
#     response = requests.get(url.format(row['ISBN']))
#     soup = BeautifulSoup(response.text, 'html.parser')  # html 파싱
    
#     try:
#         if pd.isna(title):
#             title = soup.find('a', attrs={'class' : 'gd_name'}).get_text()
    
#     except AttributeError:
#         pass
    
#     try:
#         if pd.isna(author):
#             authors = soup.find('span', attrs={'class' : 'info_auth'}).find_all('a')
#             author_list = [auth.get_text() for auth in authors]
#             author = ','.join(author_list)
            
#     except AttributeError:
#         pass
    
#     try:
#         if pd.isna(pub):
#             pub = soup.find('span', attrs={'class' : 'info_pub'}).find('a').get_text()
    
#     except AttributeError:
#         pass
    
#     try:
#         if year == -1:
#             year_str = soup.find('span', attrs={'class' : 'info_date'}).get_text()
#             year = re.findall(r'\d{4}', year_str)[0]
            
#     except AttributeError:
#         pass
    
#     return title, author, pub, year


import time
import random
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

def get_book_info(row):
    """
    도서 정보를 Yes24에서 가져오는 함수 (최적화 포함).
    """
    title = row['도서명']
    author = row['저자']
    pub = row['출판사']
    year = row['발행년도']
    isbn = row['ISBN']
    
    # Yes24 도서 검색 페이지 URL
    url = f'https://www.yes24.com/Product/Search?domain=BOOK&query={isbn}'
    max_retries = 3  # 재시도 횟수 제한
    retry_delay = random.uniform(0.5, 1)  # 재시도 시 대기 시간

    for attempt in range(max_retries):
        try:
            # 요청 전 랜덤 딜레이 추가 (0.1~0.3초)
            time.sleep(random.uniform(0.1, 0.3))
            
            response = requests.get(url, timeout=5)  # HTTP 요청
            response.raise_for_status()  # HTTP 에러 확인
            soup = BeautifulSoup(response.text, 'html.parser')  # HTML 파싱
            
            # 각 정보 파싱
            if not title:
                title_elem = soup.find('a', attrs={'class': 'gd_name'})
                title = title_elem.get_text().strip() if title_elem else None
            
            if not author:
                authors_elem = soup.find('span', attrs={'class': 'info_auth'})
                authors = authors_elem.find_all('a') if authors_elem else []
                author = ', '.join([auth.get_text().strip() for auth in authors]) if authors else None
            
            if not pub:
                pub_elem = soup.find('span', attrs={'class': 'info_pub'})
                pub = pub_elem.find('a').get_text().strip() if pub_elem else None
            
            if year == -1:
                year_elem = soup.find('span', attrs={'class': 'info_date'})
                year_str = year_elem.get_text().strip() if year_elem else None
                year = int(re.findall(r'\d{4}', year_str)[0]) if year_str else -1

            # 성공적으로 데이터를 가져오면 반환
            return title, author, pub, year
        
        except requests.exceptions.RequestException as e:
            print(f"Request error on attempt {attempt + 1} for ISBN {isbn}: {e}")
            time.sleep(retry_delay)  # 재시도 전 대기
            
        except Exception as e:
            print(f"Error processing ISBN {isbn}: {e}")
            break  # 예외 발생 시 루프 종료
    
    # 실패 시 None 반환
    return title, author, pub, year
````

````python
from concurrent.futures import ThreadPoolExecutor

# 병렬 처리 함수
def parallel_apply(df, func, max_workers=5):
    """
    DataFrame에 병렬 처리를 적용하는 함수
    """
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        results = list(executor.map(func, [row for _, row in df.iterrows()]))
    return pd.DataFrame(results)

# 실행
update_df = parallel_apply(ns_book1[na_rows], get_book_info, max_workers=5)
update_df.head(2)
````
