```python
fig, axs = plt.subplots(2)
axs[0].scatter(df['도서권수'], df['대출건수'], alpha=0.1)
axs[1].hist(df['대출건수'], bins=100)
axs[1].set_yscale('log')
fig.show()
```

```python
fig, axs = plt.subplots(2, figsize=(6,8))
axs[0].scatter(df['도서권수'], df['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')
axs[1].hist(df['대출건수'], bins=100)
axs[1].set_title('hist plot')
axs[1].set_yscale('log')

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 4))

# scatter plot
axs[0, 1].scatter(df['도서권수'], df['대출건수'], alpha=0.1)
axs[0, 1].set_title('scatter plot')
axs[0, 1].set_xlabel('number of books')
axs[0, 1].set_ylabel('borrow count')

# hist plot
axs[1, 0].hist(df['대출건수'], bins=100)
axs[1, 0].set_title('hist plot')
axs[1, 0].set_yscale('log')
axs[1, 0].set_xlabel('borrow count')
axs[1, 0].set_ylabel('frequency')

plt.show()
```

```python
plt.rcParams['axes.unicode_minus'] = False
# 3차원 배열처럼 서브플롯 구성
figs = np.empty((2, 2, 2), dtype=object)  # 2x2x2 서브플롯 배열

for i in range(2):
    for j in range(2):
        for k in range(2):
            fig, ax = plt.subplots()  # 개별 플롯 생성
            ax.plot([i, j, k], [k, j, i])  # 간단한 데이터
            ax.set_title(f"Subplot {i}-{j}-{k}")
            figs[i, j, k] = ax

plt.show()
```
