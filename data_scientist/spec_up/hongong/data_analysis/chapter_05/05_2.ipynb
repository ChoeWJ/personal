```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

plt.rc('font', family='AppleGothic')

print(plt.rcParams['figure.figsize'])
print(plt.rcParams['figure.dpi'])
```

```python
df = pd.read_csv("data/ns_book.csv", low_memory=False)

df.head()
```

```python
count_by_year = df['발행년도'].value_counts()
count_by_year
```

```python

count_by_year = count_by_year.sort_index()
```

```python
count_by_year = count_by_year[count_by_year.index <= 2030]
count_by_year
```

```python
df
```

```python
def kdc_1st_char(no):
    if no is np.nan:
        return '-1'
    return no[0]

count_by_subject = df['주제분류번호'].apply(kdc_1st_char).value_counts()
count_by_subject
```

```python
plt.plot(count_by_year.index, count_by_year.values)
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, marker='.', linestyle=':', color='red')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, '.:r')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, '*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year, '*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))
for idx, val in count_by_year[::5].items():
    plt.annotate(val, (idx, val))


plt.show()
```

```python
plt.plot(count_by_year, '*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))
for idx, val in count_by_year[::5].items():
    plt.annotate(
    "Point",
    xy=(1, 1),          # 화살표 끝점
    xytext=(0.5, 1.5),  # 텍스트 위치
    arrowprops=dict(
        arrowstyle="->",  # 화살표 스타일
        shrink=0.1,       # 화살표의 시작점과 끝점에서 10% 축소
        color="blue",     # 화살표 색상
        lw=2             # 화살표 두께
    ),
)


plt.show()
```

```python
plt.bar(count_by_subject.index, count_by_subject.values)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
    plt.annotate(val, (idx, val), xytext=(0, 2), textcoords="offset points")

plt.show()
```

```python
plt.bar(count_by_subject.index, count_by_subject.values, width=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
    plt.annotate(
        val, 
        (idx, val), 
        xytext=(0, 2), 
        textcoords="offset points",
        fontsize=8,
        ha='center',
        color='green',
        arrowprops=dict(
                    arrowstyle='->', 
                    color='green', 
                    lw=1.5,
                    connectionstyle='angle,angleA=0,angleB=90,rad=0'
        )
    )

plt.show()
```

```python
plt.barh(count_by_subject.index, count_by_subject.values, height=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('number of books')
plt.ylabel('subject')
for idx, val in count_by_subject.items():
    plt.annotate(
        val, 
        (val, idx), 
        xytext=(2, 0), 
        textcoords="offset points",
        fontsize=8,
        va='center',
        color='green'
    )

plt.show()
```
